```python
def plot_distribution(df, features):
    for feature in features:
        plt.figure(figsize=(8, 4))
        df[feature].plot(kind='hist', bins=30, edgecolor='black', alpha=0.7)
        plt.title(f'Distribution de {feature}')
        plt.xlabel(feature)
        plt.ylabel('Fréquence')
        plt.show()
```

```python
# numerical_features = ['year', 'kilometrage', 'horsepower', 'nbre_portes', 'cylindrée', 'Prix']
# Visualisation de la distribution des features numériques
plot_distribution(df, df.columns)
```
